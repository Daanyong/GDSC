```python
bins = [1, 10, 20, 30, 40, 50, 60, 70, 80, 90]
labels = ["10대미만", "10대", "20대", "30대", "40대", "50대", "60대", "70대", "80대"]
Clean_Ages = pd.cut(train_df["Age"], bins, labels=labels)
Clean_Ages
train_df["Clean_Ages"] = Clean_Ages
train_df[['Clean_Ages', 'Survived']].groupby(['Clean_Ages'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```
